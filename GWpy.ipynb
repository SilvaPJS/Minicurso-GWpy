# Dominio das frequências


A transformada de Fourier é uma ferramenta matemática amplamente utilizada para revelar o conteúdo de frequência de um sinal no domínio do tempo, permitindo visualizar quais frequências contêm mais energia e quais têm menos. Essencialmente, ela decompõe uma função periódica ou um sinal em suas frequências componentes. Quando aplicada a sinais discretos, como em computação, é conhecida como Transformada Discreta de Fourier (DFT).

Na DFT, a entrada é decomposta em suas componentes de frequência discretas, o que a torna fundamental para o processamento digital de sinais, incluindo filtragem e análise espectral. A fórmula da DFT para um sinal discreto ( **a** ) de comprimento ( **n** ) é dada por:

$$A_{k} = \sum_{m=0}^{n-1} a_{m} e^{-2\pi i \frac{mk}{n}}, \quad \text{para } k=0, \ldots, n-1.$$

Esta equação calcula os coeficientes ( $A_{k}$ ) que representam as amplitudes das componentes de frequência discretas do sinal.

No contexto de análise de dados de ondas gravitacionais, a DFT é usada para transformar séries temporais de sinais captados pelos detectores em representações de frequência, permitindo a identificação e estudo das características espectrais dos eventos detectados.

 Quando a entrada $a$ é um sinal no domínio do tempo e a DFT é aplicada a $a$, resultando na transformada $A$, o módulo ou valor absoluto da transformada ($|A|$) representa o espectro de amplitude, e o módulo ao quadrado ($|A|^2$) representa o espectro de potência do sinal.

A função `fft()`, presente no *GWpy*, utiliza ferramentas presentes no *NumPy* para o cálculo da DFT unidimensional de um objeto do tipo série temporal.

```python
fft = ldata.fft()
print(fft)
```

O método *fft()* foi empregado para calcular a transformada de Fourier da série temporal em questão. Assim como no NumPy, pode-se utilizar a função `abs()` para extrair a amplitude.

```python
plot = fft.abs().plot(xscale="log", yscale="log")

axfft = plot.gca()
axfft.set_xlim(10, 1500)
axfft.set_ylim(0.4e-26, 1e-19)
plot.show()
```

A utilização do gráfico em escala log-log proporciona uma melhor visualização da distribuição da amplitude das frequências ao longo do espectro de Fourier. Isso é útil para identificar quais frequências têm maior ou menor contribuição de energia no sinal analisado.

Ao utilizar a FFT para medir o componente de frequência de um sinal, a análise é baseada em um conjunto finito de dados. Assim, a transformada FFT assume que é um conjunto de dados finitos, sendo um espectro contínuo que é um período de um sinal periódico. Para essa transformada, tanto o domínio do tempo quanto o de frequência são tratados como topologias circulares, então os dois pontos finais da forma de onda no tempo são interpretados como se estivessem conectados. Quando o sinal medido é periódico e um número inteiro de períodos preenche o intervalo de tempo de aquisição, a FFT funciona bem, por corresponder a essa suposição.

No entanto, na prática, os dados geralmente não são perfeitamente periódicos, resultando em descontinuidades ou saltos abruptos nas bordas (início e fim) do conjunto de dados. Portanto, a finitude do sinal medido pode resultar em uma forma de onda truncada com características diferentes do sinal original em tempo contínuo, e essa finitude pode introduzir mudanças bruscas de transição no sinal medido. Essas transições bruscas são descontinuidades.

O espectro obtido usando uma FFT, portanto, não é o espectro real do sinal original, mas uma versão borrada, como se a energia de uma frequência vazasse para outras frequências. Esse fenômeno é conhecido como vazamento espectral, que faz com que as linhas espectrais finas se espalhem em sinais mais amplos.

É possível minimizar os efeitos de realizar uma FFT sobre um número não inteiro de ciclos usando uma técnica chamada janela (**windowing**). A janela consiste em multiplicar o registro temporal por uma janela de comprimento finito com uma amplitude que varia suavemente e gradualmente em direção a zero nas bordas. Existem vários tipos de funções janela que são possíveis de aplicar, variando conforme o sinal observado. Em muitos casos, a janela de Hann se destaca devido à sua resolução de frequência e capacidade de reduzir a dispersão espectral. Esta função de janela toca zero em ambas as extremidades, eliminando assim qualquer descontinuidade, sendo definida pela equação:


$$w(n) = 0.5 - 0.5\cos\left(\frac{2\pi n}{M-1}\right), \quad \text{onde} \quad 0 \leq n \leq M-1.$$

Por essas características, a janela de Hann, também conhecida como *Cosine Bell*, é comumente usada para suavizar valores.

windowing_shg (1).jpg



 Para aplicar uma janela de Hann em um conjunto de dados, é possível usar a função `get_window` do módulo `scipy.signal`, como mostrado a seguir:


```python
from scipy.signal import get_window
window = get_window('hann', ldata.size)
lwin = ldata * window
```

Note que o `size` é utilizado para criar uma janela de **Hann** com a mesma quantidade de pontos do sinal, garantindo que a janela e o sinal tenham o mesmo comprimento e permitindo a multiplicação ponto a ponto. A multiplicação da janela só pode ser realizada se ambos tiverem o mesmo comprimento. Após suavizar as descontinuidades dos dados por meio desse processo, é possível aplicar a FFT, resultando na obtenção da estimativa da FFT e no cálculo do seu módulo. Esse procedimento retorna o espectro de amplitude para um único sinal.



```python
fftamp = lwin.fft().abs()
plot = fftamp.plot(xscale="log", yscale="log")

limfftamp = plot.gca()
limfftamp.set_xlim(10, 1500)
limfftamp.set_ylim(0.4e-26, 1e-19)

plot.show()
```

 Resultando em uma densidade espectral de amplitude de um detector de ondas gravitacionais mais próximo do real. Como discutido anteriormente, assim como o `gwosc` e o `numpy`, o `Scipy` também faz parte do pacote de software `GWpy`, podendo utilizar suas ferramentas, criar funções e mesclar funcionalidades.


# Amplitude de Densidade Espectral (ASD)

O pacote GWpy também possibilita realzar o cálculo da  densidade espectral de potência, a densidade espectral de amplitude e a curva de sensibilidade, para melhor caracterização dos dados observados. A densidade espectral de potência (PSD) é definida como:


$$   S_{n}(f)=\displaystyle \lim_{T \to \infty
} \frac{2}{T} \left| \int_{-T/2}^{T/2}n(t)e^{2\pi i ft}dt\right|^{2}\,,$$


onde $n(t)$ é a *noise fluctuations* dos dados, $f$ é a frequência linear e $t$ é o tempo, observado em um  intervalo de $-T/2 < t < T/2$. A densidade espectral é uma medida das flutuações médias quadráticas do ruído em uma determinada frequência. O ruído nos detectores LIGO e Virgo não são estacionários. No entanto, ao medir a densidade espectral ao longo de um intervalo de tempo curto o suficiente, é possível aproximar o ruído como estacionário. O intervalo de tempo escolhido também deve ser longo o suficiente para formar uma estimativa precisa da densidade espectral.

A PSD é medida em unidades de $Hz^{-1}$, o que pode ser interpretado como a quantidade de variação em cada intervalo de frequência. Tomando a raiz quadrada da PSD encontra-se a densidade espectral de amplitude. Da mesma forma, as unidades da ASDs são $1/\sqrt{Hz}$.  Frequências onde a ASD é menor, indicam onde o detector é relativamente sensível, enquanto valores grandes indicam muito ruído e uma menor sensibilidade. Portanto, a  ASD reflete a curva de sensibilidade do instrumento durante o período de observação.

O cálculo da ASD é realizado por meio da combinação de estimativas de FFT em segmentos de dados sobrepostos. Ou seja, o conjunto de dados é dividido em vários segmentos menores, com intervalos que se sobrepõem ao início ou ao final do segmento subsequente. Após a divisão, é realizada a atenuação das descontinuidades por meio da janela de Hann. Após calcular as FFTs para esses segmentos, elas são combinadas conforme o método de **Welch**, que aplica uma média ponderada, dando mais importância aos segmentos centrais. Uma alternativa ao método de Welch é o cálculo por meio das medianas, denominado **median**. Em outras palavras, o método median adota a mediana das estimativas de FFT para mitigar flutuações e valores discrepantes. Finalmente, calcula-se a raiz quadrada do resultado, obtendo-se a ASD.

Com o `GWpy`, é possível calcular a ASD usando o comando `asd()`. Como mencionado anteriormente, a ASD pode ser calculada de duas maneiras, utilizando os métodos `Welch` ou `Median`. O comando `asd()` aceita ambos como parâmetros. No exemplo a seguir, optou-se pelo método `Median`, que é utilizado como padrão, no entanto, ambos os métodos produzem resultados adequados para a análise de sinais de ondas gravitacionais. O parâmetro `fftlength` é utilizado para alterar o tamanho das janelas individuais usadas para calcular a análise de frequência do sinal.


```python
asd = ldata.asd(fftlength=3)
plot = asd.plot()

ax = plot.gca()
ax.set_title('ASD')
ax.set_ylabel(r'Strain [$1/\sqrt{\mathrm{Hz}}$]')
ax.set(xlim=(15, 1500), ylim=(0.4e-23, 1e-20))
```

```python
# para Hanford data
hdata2 = TimeSeries.fetch_open_data('H1', *segment, cache=True)
hasd2 = hdata2.asd(fftlength=3)

ax.plot(hasd2, label='LIGO-Hanford', color='gwpy:ligo-hanford')

lline = ax.lines[0]
lline.set_color('gwpy:ligo-livingston')
lline.set_label('LIGO-Livingston')

ax.set_ylabel(r'Strain noise [$1/\sqrt{\mathrm{Hz}}$]')
ax.legend()
plot
```

O gráfico apresenta a densidade espectral de amplitude (ASD) do ruído de deformação nos detectores H1 e L1 do Advanced LIGO durante um segmento temporal em torno do evento GW150914. Note que o gráfico cobre a faixa de frequência entre 15 Hz e 1500 Hz.

Abaixo de 10 Hz, os dados não estão calibrados devido à elevada quantidade de ruído, o que impede a detecção de deformações causadas por ondas gravitacionais abaixo dessa faixa de frequência.

Observa-se a presença de fortes linhas espectrais nos dados, todas de origem instrumental. Algumas dessas linhas são instrumentais, como as ressonâncias de suspensão dos espelhos em torno de 500 Hz e seus harmônicos, além de linhas de calibração e controle.

Importante ressaltar que o sinal do evento não é visível neste gráfico, pois é relativamente fraco e dura menos de um segundo, enquanto este gráfico representa uma média de 200 segundos de dados. Assim, o gráfico é dominado pelo ruído instrumental. Em suma, a ASD reflete a curva de sensibilidade do instrumento no tempo de observação.




# Q-transform com o GWpy

Uma das ferramentas mais úteis para filtrar e visualizar recursos de curta duração na *TimeSeries* é a *Q-transform*. A *Q-transform* é utilizada para criar uma representação tempo-frequência dos dados, permitindo a identificação de características em várias frequências e sua evolução temporal em intervalos muito curtos, sem depender de um conhecimento prévio sobre a forma do sinal. A *Q-transform* oferece uma análise detalhada das mudanças temporais nas características de frequência do sinal, tornando-se uma ferramenta valiosa para a investigação de eventos astrofísicos e para a detecção de padrões complexos. Isso é particularmente útil em cenários onde a natureza exata do sinal é desconhecida ou pode variar ao longo do tempo.

A *Q-transform* discreta é definida da seguinte forma:
$$
x[m,k] = \sum_{n=0}^{N-1} x[n] w[n - m,k] e^{-i2\pi nk/N}\,,
$$
onde $x[m,k]$ representa o resultado da *Q-transform* discreta para o instante de tempo $m$ e a frequência $k$, apresentando a amplitude do sinal na frequência $k$ naquele instante específico. O termo $x[n]$ é o valor do sinal de entrada no ponto $n$, sendo o dado original que será analisado. A função janela $w[n - m,k]$ é deslocada por $m$ unidades e ajustada para a frequência $k$, sendo utilizada para isolar uma parte específica do sinal ao redor do ponto $m$, reduzindo descontinuidades e minimizando efeitos de borda na análise de *Fourier*. Por fim, $N$ representa o número total de pontos de dados no sinal, definindo o tamanho do sinal que está sendo analisado.


O pacote de software *GWpy* oferece a possibilidade de calcular a *Q-transform* através do comando a `q_transform`. Para a aplicação da da Q-transform, assim como no exemplo anterior, é utilizado um seguimento temporal. Para o seguinte exemplo, é utilizado um intervalo simétrico em torno do tempo GPS do evento GW150914, gerando um segmento de dados que abrange 16 segundos antes e depois do evento, criando uma janela de 32 segundos ao redor do evento. Em seguida, é utilizado o identificador H1 para buscar os dados do observatório LIGO-Hanford e utilizá-los na aplicação do método.

```python
from gwosc.datasets import event_gps
from gwpy.timeseries import TimeSeries

gps = event_gps('GW150914')
print("GW150914 GPS:", gps)

segment = (int(gps) - 16, int(gps) + 16)
hdata = TimeSeries.fetch_open_data('H1', *segment, verbose=True, cache=True)
```

Utilizando o comando `q_transform` do objeto *TimeSeries* para calcular a Q-transform dos dados.

```python
hq = hdata.q_transform()
plot = hq.plot()
plot.colorbar(label="Normalised energy")
```

O gráfico gerado permite uma análise visual das características do sinal em diferentes frequências ao longo do tempo, facilitando a identificação de eventos transitórios e a análise de potenciais sinais de ondas gravitacionais. A ferramenta GWpy permite ajustar as escalas e ampliar áreas específicas do gráfico, possibilitando uma investigação precisa dos dados, essencial para a pesquisa em astronomia de ondas gravitacionais.

Destaca-se a presença de um sinal significativo no intervalo entre 15 e 18 segundos. Utilizando as próprias ferramentas presentes no GWpy é possível ampliar e alterar as escalas em torno desse sinal.

```python
hq = hdata.q_transform(frange=(30, 500), outseg=(gps-0.2,gps+0.1))
plot = hq.plot()
ax = plot.gca()
ax.set_epoch(gps)
ax.set_yscale('log')
ax.colorbar(label="Normalised energy")
```

Neste exemplo, foi aplicado novamente a Q-transform aos dados, mas restringindo a faixa de frequência (`frange`) entre 30 e 500 Hz para focar em uma área de interesse mais específica. Também é definido um intervalo de saída (`outseg`) de 0.3 segundos em torno *merger* do evento GW150914.

Ao aplicar a Q-transform aos dados de ondas gravitacionais, é possível identificar como as frequências se modificam ao longo do tempo e como diferentes componentes de frequência podem estar presentes em momentos distintos. Dessa forma é possível distinguir variações naturais do sinal de possíveis eventos astrofísicos, como fusões de objetos compactos. Também é possível observar uma trilha clara em cerca de $-0.06$ segundos, aumentando até alcançar o pico em $0,03$, onde também é encontrada a região mais energética. Como já é conhecido, o sinal se trata de um sistema binárias de buracos negros, onde o pico apresentado expressa o momento da fusão dos objetos. Portanto o segmento se trata da fase de coalescência.
